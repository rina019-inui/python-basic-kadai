```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
      if  self.age >= 20:
        return f"{self.name}は大人です。"
      else:
        return f"{self.name}は大人ではありません。"

person1 = Human("太郎", 25)
person2 = Human("花子", 15)
person3 = Human("健太", 30)

people_rist = [person1, person2, person3]

for person in people_rist:
    print(person.check_adult())
```
